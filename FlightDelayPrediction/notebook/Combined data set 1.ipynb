```python
import pandas as pd
import glob
import numpy as np
import os

# List all files in the /content directory
files_in_content = os.listdir('/content')
print("Files in /content directory:", files_in_content)

# Set the path to your CSV files
path = '/content/*.csv'  # Adjust this path based on your file location
all_files = glob.glob(path)
print("Detected CSV files:", all_files)  # Check if files are detected

# Initialize an empty list to store DataFrames
dfs = []

# Loop through each file, read it with error handling, sample 100,000 rows, and append it to the list
for file in all_files:
    print(f'Reading {file}')  # Check each file being processed
    try:
        df = pd.read_csv(file, on_bad_lines='skip')  # Skip bad lines
        if df.empty:
            print(f'Warning: {file} is empty or failed to load')
            continue
        print(df.head())  # Print the first few rows to confirm data is read
        if len(df) > 100000:
            df = df.sample(n=100000, random_state=42)  # Randomly sample 100,000 rows
        dfs.append(df)
    except pd.errors.ParserError as e:
        print(f'Error reading {file}: {e}')
        continue

# Concatenate all DataFrames in the list into a single DataFrame by matching column names
if dfs:
    combined_df = pd.concat(dfs, ignore_index=True, sort=False)
    # Save the combined DataFrame to a new CSV file
    combined_df.to_csv('/content/combined_sampled_by_columns.csv', index=False)
    print("Combined data saved to /content/combined_sampled_by_columns.csv")
else:
    print("No DataFrames to concatenate. Please check the files.")
```
